# Analyzation Between Region & CTSA

```python
rectsa=stud_data['Region'].groupby(stud_data['CTSA']).count()
```

```python
rectsa
```

# Diagramatic Representation Of Above Analysis

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(rectsa)
```

# Analyzation Between The Schools

```python
column_sums = df[['CTSA', 'GOVT', 'GOVTAIDED', 'INDEPENDENT', 'JNV', 'KV']].sum()
```

```python
column_sums
```

# Diagramatic Representation Of Above Analysis

```python
plt.figure(figsize=(8, 6))
plt.bar(column_sums.index, column_sums.values)
plt.xlabel('Schools')
plt.xticks(rotation = 45,ha='right')
plt.ylabel('Sum')
plt.title('Different school pass rate')
plt.show()
```

# Analyzation Based on Region

```python
rates_variations = df.loc[:, ['Region', 'Status','sum_column']].groupby(['Region', 'Status']).max().reset_index()
rates_variations['sum_column'] = rates_variations['sum_column'].astype(int)
rates_variations.columns = ['Region', 'Status', 'sum_column']
```

```python
rates_variations
```

# Diagramatic Representation Of Above Analysis

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Region', y='sum_column', data=rates_variations, hue='Status', palette='Blues_d')
plt.title('Variations in pass rates across different regions')

plt.xticks(rotation = 45,ha='right')
plt.xlabel('Regions')
plt.ylabel('Rate Quantity')

plt.legend(title='Rates', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Conclusion

**By analysing the above data,we can clearly understand that,INDEPENDENT School got the highest number of passing rates,Region wise Delhi West got highest number of passing rates.**
